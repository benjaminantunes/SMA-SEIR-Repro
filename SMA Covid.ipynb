```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import pandas as pd
import csv
from statistics import mean,variance


NB_REPLICATION = 4






for replication in range(0,NB_REPLICATION):

    listNumIteration = []
    listNbSusceptible = []
    listNbExposed = []
    listNbInfected = []
    listNbRecovered = []


    f = open("log"+str(replication)+".txt", "r")
    fileLines = f.readlines()
    i = 0
    for line in fileLines:
        if("##" not in line):
            if(i == 0):
                listNumIteration.append(int(line))
                i = i + 1
            elif(i==1):
                listNbSusceptible.append(int(line))
                i = i + 1
            elif(i==2):
                listNbExposed.append(int(line))
                i = i + 1
            elif(i==3):
                listNbInfected.append(int(line))
                i = i + 1
            elif(i==4):
                listNbRecovered.append(int(line))
                i = i + 1
        else:
            i = 0

    f.close()



    nbHumainsTotal = listNbSusceptible[0] + listNbInfected[0]
    print(nbHumainsTotal)

    listNbSusceptible = [i/nbHumainsTotal for i in listNbSusceptible]

    listNbExposed = [i/nbHumainsTotal for i in listNbExposed]
    listNbInfected = [i/nbHumainsTotal for i in listNbInfected]
    listNbRecovered = [i/nbHumainsTotal for i in listNbRecovered]

    plt.figure(str(replication))
    print("Courbe des résultats sur la réplication num : " +str(replication))
    plt.plot(listNumIteration, listNbSusceptible, label="susceptible")  # Matplotlib plot.
    plt.plot(listNumIteration, listNbExposed, label="exposed")  # Matplotlib plot.
    plt.plot(listNumIteration, listNbInfected, label="infected")  # Matplotlib plot.
    plt.plot(listNumIteration, listNbRecovered, label="recovered")  # Matplotlib plot.
    plt.xlabel('Time in days')  # Add an x-label to the axes.
    plt.ylabel('Ratio of agents in each state')  # Add a y-label to the axes.
    plt.title("S, E, I and R agents population over time")  # Add a title to the axes.
    plt.legend(loc="best")
    plt.savefig("resulats_replication"+str(replication)+".png")
    string = "resulats_replication"+str(replication)+".csv"
    with open(string, "w") as fCsv:
        write = csv.writer(fCsv)
        write.writerow(listNbSusceptible)
        write.writerow(listNbExposed)
        write.writerow(listNbInfected)
        write.writerow(listNbRecovered)
```
